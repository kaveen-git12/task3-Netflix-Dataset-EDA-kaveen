# Netflix Dataset Exploratory Data Analysis (EDA)

This notebook performs Exploratory Data Analysis on the Netflix dataset to uncover patterns and trends.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visualizations
%matplotlib inline
sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load the dataset
df = pd.read_csv('netflix_titles.csv')
df.head()
```

## 1. Summary Statistics


```python
# Dataset Shape
print(f'Dataset Shape: {df.shape}')
df.shape
```

```python
# Missing values
df.isnull().sum()
```

```python
# Data types and basic info
df.info()
```

```python
# Basic statistics
df.describe(include='all')
```

### Insight:
The dataset contains 7787 rows and 12 columns. There are some missing values, notably in the `director`, `cast`, and `country` columns. Most features are categorical (object types), while `release_year` is numerical.


## 2. Correlation Analysis


```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

### Insight:
Usually, there is a weak correlation in the Netflix dataset between numerical variables (like release_year). The heatmap confirms that there are no strong linear relationships between the available numerical features.


## 3. Trend Identification


### 3.1 Content Type (Movie vs TV Show)


```python
plt.figure(figsize=(8, 5))
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Content Type: Movie vs TV Show')
plt.show()
```

### Insight:
Netflix has significantly more Movies than TV Shows in its catalog. This suggests a stronger focus on feature-length content.


### 3.2 Top Countries Producing Content


```python
plt.figure(figsize=(10, 6))
df['country'].value_counts().head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Content Producing Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Insight:
The United States is the leading producer of content on Netflix, followed by India and the United Kingdom. This aligns with the major global entertainment markets.


### 3.3 Content Added Over Time


```python
# Extract the year content was added
df['year_added'] = df['date_added'].str.strip().str[-4:]
# Convert to numeric, errors='coerce' will handle any missing or malformed data
df['year_added'] = pd.to_numeric(df['year_added'], errors='coerce')

plt.figure(figsize=(10, 6))
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='purple', linewidth=2)
plt.title('Content Added to Netflix Over Time')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

### Insight:
There was an exponential growth in content addition starting around 2015, peaking around 2019/2020. This corresponds to Netflix's major global expansion and push for original content.


### 3.4 Genre Distribution


```python
plt.figure(figsize=(10, 8))
df['listed_in'].value_counts().head(10).plot(kind='barh', color='coral', edgecolor='black')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.gca().invert_yaxis() # Highest at the top
plt.show()
```

### Insight:
'Documentaries', 'Stand-Up Comedy', and 'Dramas, International Movies' are among the most common genres. Netflix clearly invests heavily in international content and documentaries.


### 3.5 Ratings Distribution


```python
plt.figure(figsize=(12, 6))
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index, palette='viridis')
plt.title('Content Ratings Distribution')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

### Insight:
TV-MA (Mature Audiences) and TV-14 are the most prominent ratings, indicating that a significant portion of Netflix's content is targeted towards adult and older teen audiences rather than young children.
